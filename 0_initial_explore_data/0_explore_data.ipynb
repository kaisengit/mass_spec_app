# IMPORTS

```python
import pandas as pd
import sweetviz as sv
import os.path
```

# INPUT PATHS

```python
PATH_DATA = "./data/"
PATH_COMPOUNDS = os.path.join(PATH_DATA, "compounds.xlsx")
PATH_COMPOUNDS_MEASURED = os.path.join(PATH_DATA, "measured-compounds.xlsx")
PATH_ADDUCTS = os.path.join(PATH_DATA, "adducts.json")
```

# READ DATAFRAMES

```python
df_compounds = pd.read_excel(PATH_COMPOUNDS)
df_compounds_measured = pd.read_excel(PATH_COMPOUNDS_MEASURED)
df_adducts = pd.read_json(PATH_ADDUCTS)
```
